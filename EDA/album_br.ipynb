```python
#long format as album can have multiple artists or only one
#one artist per line
import pandas as pd
import json

with open("../data/albums/BR/2025-10-09.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.json_normalize(data, sep="_", max_level=1)

df = df.explode("artists")

artist_cols = pd.json_normalize(df['artists']).add_prefix('artist_')
df = df.drop(columns=['artists']).reset_index(drop=True).join(artist_cols)

df.head()
```
